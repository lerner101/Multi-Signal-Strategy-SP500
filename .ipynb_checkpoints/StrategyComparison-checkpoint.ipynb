## Plots for Each Strategy
Here we plot the signal overlays on the price chart. We chooose one ticker for each strategy here, `AAPL`.

### Signal Overlay on Price Chart

```python
def plot_signal_overlay(res_df, ticker: str, title_prefix=""):
    price = res_df["Price"][ticker]
    buys  = res_df["ExecSignal"][ticker] == 1
    sells = res_df["ExecSignal"][ticker] == -1

    plt.figure(figsize=(10,4))
    plt.plot(price.index, price.values, label=f"{ticker} Close")
    
    # markers
    plt.scatter(price.index[buys],  price[buys],  marker="^", s=40, label="Buy (exec)", zorder=3)
    plt.scatter(price.index[sells], price[sells], marker="v", s=40, label="Sell (exec)", zorder=3)
    plt.title(f"{title_prefix} Signal Overlay â€” {ticker}")
    plt.xlabel("Date"); plt.ylabel("Price")
    plt.legend(); plt.show()
```

```python
plot_signal_overlay(res_ma, "AAPL", title_prefix="MA(20/50)")
```

```python
plot_signal_overlay(res_macd, "AAPL", title_prefix="MACD")
```

```python
plot_signal_overlay(res_rsi, "AAPL", title_prefix="RSI")
```

```python
plot_signal_overlay(res_vb, "AAPL", title_prefix="Vol Breakout")
```

```python
plot_signal_overlay(res_b, "AAPL", title_prefix="Benchmark")
```

### Plot for Holdings, Cash and Total Assets

```python
def plot_holdings_cash_assets(res_df, title_prefix=""):
    fig, ax = plt.subplots(2, 1, figsize=(10,8), sharex=True)

    # Total holdings, just the sum across tickers
    total_shares = res_df["Holdings"].sum(axis=1)
    ax[0].plot(total_shares.index, total_shares.values)
    ax[0].set_title(f"{title_prefix} Total Shares Held")
    ax[0].set_ylabel("Shares")

    equity = res_df[["Cash", "TotalAssets"]]
    ax[1].plot(equity.index, equity["TotalAssets"], label="Total Assets")
    ax[1].plot(equity.index, equity["Cash"], label="Cash", linestyle="--")
    ax[1].set_title(f"{title_prefix} Cash & Total Assets")
    ax[1].set_xlabel("Date"); ax[1].set_ylabel("USD")
    ax[1].legend()

    # show plain numbers
    for a in ax: a.ticklabel_format(axis="y", style="plain", useOffset=False)
    plt.show()
```

```python
plot_holdings_cash_assets(res_ma, title_prefix="MA(20/50)")
```

```python
plot_holdings_cash_assets(res_macd, title_prefix="MACD")
```

```python
plot_holdings_cash_assets(res_rsi, title_prefix="RSI")
```

```python
plot_holdings_cash_assets(res_vb, title_prefix="Vol breakoout")
```

```python
plot_holdings_cash_assets(res_b, title_prefix="Benchmark")
```

## Comparing CumPnL of Each Strategy

```python
def run_strategy(strategy_obj, prices, initial_cash=initial_cash):
    """
    Just a helper to run strategies again 
    """
    sim = Simulator(strategy=strategy_obj, prices=prices, initial_cash=initial_cash)
    res = sim.run()
    return res


def equity_return_series(res_df: pd.DataFrame) -> pd.Series:
    nav = res_df["TotalAssets"] / res_df["TotalAssets"].iloc[0] - 1.0
    nav.name = res_df.attrs.get("strategy_name", "Strategy")
    return nav

def compare_cum_pnl(prices_matrix: pd.DataFrame, strategies: list):
    navs = []
    for strat in strategies:
        res = run_strategy(strat, prices_matrix)
        navs.append(equity_return_series(res))
    navs = pd.concat(navs, axis=1)

    plt.figure(figsize=(10,5))
    for c in navs.columns:
        plt.plot(navs.index, navs[c], label=c)
    plt.title("Cumulative PnL (Return) Comparison")
    plt.xlabel("Date"); plt.ylabel("Return")
    plt.legend(); plt.show()
    return navs

navs_small = compare_cum_pnl(
    P,
    [
        MA,
        MACD,
        RSI,
        VolBreakout,
        B
    ]
)
navs_small.tail()
```

## Reflection
The strategies behave quite differently under the rules given in the assigment. MA(20/50) signal rides trends, emitting long streaks of buys that steadily build exposure.RSI buys dips and sells rips, creating moderate mean-reverting exposure. MACD flips frequently, so with a one-share-per-day cap it trades a lot but rarely builds meaningful size. The volatility breakout, defined as daily return exceeding the 20-day standard deviation of returns, fires on big moves and builds exposure fast during leg ups.

The execution constraints drive the ranking. The one-share-per-day rule throttles position size, the no-short rule means sell signals only scale out of longs and we dont really use it to take advantage of short opportunities. 

In this period of time, MA(20/50) performs best because it accumulates the most size, and the market regime is generally an uptrend. The volatility breakout follows. RSI is steady but we can see that it lags in strong trends. Finally MACD is weakest. Overall, under the constraints given, trend-following wins and turnover-heavy signals are structurally disadvantaged in this market regime from 2015 to 2025.

Comparing to benchmark, which is a simple buy on day 1 and hold strategy, strategies that keep adding in good regimes beat it.
- MA(20/50) and Vol Breakout outperform because they systematically accumulate more shares during uptrends.
- RSI is only slightly better than the benchmark since it buys dips but also scales out, so exposure grows slower.
- MACD often lags. Frequent flips prevent size from building.

Conclusion: Against a buy-and-hold yardstick, trend-followers win, RSI is middling, and high-turnover MACD is disadvantaged by the daily unit cap.
